#### Outlier Detection

```python
last_data
```

```python
Q1 = last_data.drop(['Participant_Stimuli','ResponseCode','TypeCode','AgeCode','SexCode','participant','stimuli'],axis=1).quantile(0.25)
Q3 = last_data.drop(['Participant_Stimuli','ResponseCode','TypeCode','AgeCode','SexCode','participant','stimuli'],axis=1).quantile(0.75)
IQR = Q3 - Q1
```

```python
IQR
```

```python
((last_data.drop(['Participant_Stimuli','ResponseCode','TypeCode','AgeCode','SexCode','participant','stimuli'],axis=1) < (Q1 - 1.5 * IQR)) | (last_data.drop(['Participant_Stimuli','ResponseCode','TypeCode','AgeCode','SexCode','participant','stimuli'],axis=1) > (Q3 + 1.5 * IQR))).sum()
```

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```

```python
wo_outliers = last_data
```

```python
for i in range(1,65):
    wo_outliers = remove_outlier(wo_outliers,list(wo_outliers.columns)[i])
```
